```python
import pandas as pd
data = pd.read_csv("new.csv",encoding='ISO-8859-1')
# data
dt = {}
camps = list(data['Campaign'])
camp = {}
for i in camps:
    if(i not in camp):
        camp[i] = 0
    else:
        camp[i] += 1
b = [1640, 673, 658, 549, 465]
a = [] #storing the top 5 campaigns
for i in camp:
    if(camp[i] in b):
        a.append(i)
        b.remove(camp[i])
for i in a:
    dt[i] = []
    
camps = list(data['Interested_in'])
c = set(camps)
ints = {} #storing the unique intersted_in
for i in c:
    temp = i.split(',')
    for j in temp:
        if(len(j)>1):
            ints[j.strip(' ')] = 0
# ints
# dt
for i in dt:
    dt[i] = dict(ints)
# a   storing the top 5 campaigns
dt

cmp = list(data['Campaign'])
intin = list(data['Interested_in'])

n = len(intin)

for i in range(n):
    if(cmp[i] in a):
        for j in ints:
            if(j in intin[i]):
                dt[cmp[i]][j] += 1
                
```
