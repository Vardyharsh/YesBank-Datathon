```python
df = pd.read_csv("new2.csv", encoding="ISO-8859-1")
df = df[['Interested_in','City','Campaign','Disposition','Disp']]
```

```python
# print(df.shape)
```

```python
series= pd.isnull(df["Interested_in"])
df = df[series]
```

```python
bool_series = pd.notnull(df['Campaign'])
# df[bool_series]
```

```python
df = df[['Interested_in','Campaign']]
# df
```

```python
print(df.shape)
df=df.dropna(how='all')
# df.Campaign.count()
```

```python
lt = list(df['Campaign'])
b = {}
for i in lt:
    if(i not in b):
        b[i] = 1
    else:
        b[i] += 1
        
k = list(b.values())
k.sort()

slices = k[-12:]
activities = []
for i in b:
    if(b[i] in slices):
        activities.append(i)
import matplotlib.pyplot as plt

plt.pie(slices, labels =activities,startangle =90, shadow = True, autopct = '%1.1f%%')
plt.title("Top 12 Ineffective Campaigns\n")
```

```python
import matplotlib.pyplot as plt

s1 = sum(slices)
s2 = sum(k) - s1

plt.pie([s1,s2], labels =['Top 12 Ineffective Campaigns', 'Rest of Ineffective Campaign'],startangle =90, shadow = True, autopct = '%1.1f%%')
plt.title("Ineffective Campaigns Division\n")
```
